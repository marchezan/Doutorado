```python
import matplotlib.pyplot as plt
import numpy as np

# Model names and accuracy values
models = ["KNN Car 1", "KNN Car 2", "ANN Car 1", "ANN Car 2", "Gradient Car 1", "Gradient Car 2"]
accuracies = [0.89, 0.94, 0.88, 0.96, 0.92, 0.97]

# Create bar chart
plt.figure(figsize=(10, 5))
plt.bar(models, accuracies, color=['blue', 'blue', 'green', 'green', 'red', 'red'])

# Labels and title
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Comparison of Model Accuracies")
plt.ylim(0.85, 1.0)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show values on top of bars
for i, v in enumerate(accuracies):
    plt.text(i, v + 0.005, f"{v:.2f}", ha='center', fontsize=10)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados extrados do arquivo
models = ["KNN Car 1", "KNN Car 2", "ANN Car 1", "ANN Car 2", "Gradient Car 1", "Gradient Car 2"]

precision = [0.88, 0.95, 0.88, 0.96, 0.92, 0.97]
recall = [0.88, 0.94, 0.88, 0.96, 0.92, 0.97]
f1_score = [0.88, 0.95, 0.88, 0.96, 0.92, 0.97]

# Criar grfico de barras agrupadas
x = np.arange(len(models))
width = 0.25

plt.figure(figsize=(12, 6))
plt.bar(x - width, precision, width, label="Precision", color='blue')
plt.bar(x, recall, width, label="Recall", color='green')
plt.bar(x + width, f1_score, width, label="F1-Score", color='red')

# Rtulos e ttulo
plt.xlabel("Models")
plt.ylabel("Scores")
plt.title("Comparison of Precision, Recall, and F1-Score")
plt.xticks(x, models, rotation=45)
plt.ylim(0.75, 1.0)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar valores no topo das barras
for i in range(len(models)):
    plt.text(x[i] - width, precision[i] + 0.005, f"{precision[i]:.2f}", ha='center', fontsize=9)
    plt.text(x[i], recall[i] + 0.005, f"{recall[i]:.2f}", ha='center', fontsize=9)
    plt.text(x[i] + width, f1_score[i] + 0.005, f"{f1_score[i]:.2f}", ha='center', fontsize=9)

# Mostrar grfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
models = ["KNN Car 1", "KNN Car 2", "ANN Car 1", "ANN Car 2", "Gradient Car 1", "Gradient Car 2"]
precision = [0.88, 0.95, 0.88, 0.96, 0.92, 0.97]
recall = [0.88, 0.94, 0.88, 0.96, 0.92, 0.97]
f1_score = [0.88, 0.95, 0.88, 0.96, 0.92, 0.97]
accuracy = [0.89, 0.94, 0.88, 0.96, 0.92, 0.97]

# Criar grfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(models, precision, marker='o', linestyle='-', label="Precision", color='blue')
plt.plot(models, recall, marker='s', linestyle='-', label="Recall", color='green')
plt.plot(models, f1_score, marker='D', linestyle='-', label="F1-Score", color='red')
plt.plot(models, accuracy, marker='^', linestyle='-', label="Accuracy", color='purple')

# Adicionar rtulos e ttulo
plt.xlabel("Models")
plt.ylabel("Scores")
plt.title("Comparison of Model Performance (Line Chart)")
plt.xticks(rotation=45)
plt.ylim(0.75, 1.0)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar grfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Criar DataFrame com os dados
data = {
    "Model": ["KNN Car 1", "KNN Car 2", "ANN Car 1", "ANN Car 2", "Gradient Car 1", "Gradient Car 2"],
    "Precision": [0.88, 0.95, 0.88, 0.96, 0.92, 0.97],
    "Recall": [0.88, 0.94, 0.88, 0.96, 0.92, 0.97],
    "F1-Score": [0.88, 0.95, 0.88, 0.96, 0.92, 0.97],
    "Accuracy": [0.89, 0.94, 0.88, 0.96, 0.92, 0.97]
}

df = pd.DataFrame(data)
df.set_index("Model", inplace=True)

# Criar o heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Adicionar ttulo
plt.title("Heatmap of Model Performance Metrics")

# Mostrar grfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Criar DataFrame com os dados
data = {
    "Model": ["KNN Car 1", "KNN Car 2", "ANN Car 1", "ANN Car 2", "Gradient Car 1", "Gradient Car 2"],
    "Precision": [0.88, 0.95, 0.88, 0.96, 0.92, 0.97],
    "Recall": [0.88, 0.94, 0.88, 0.96, 0.92, 0.97],
    "F1-Score": [0.88, 0.95, 0.88, 0.96, 0.92, 0.97],
    "Accuracy": [0.89, 0.94, 0.88, 0.96, 0.92, 0.97]
}

df = pd.DataFrame(data)

# Converter para formato longo para Seaborn
df_melted = df.melt(id_vars=["Model"], var_name="Metric", value_name="Score")

# Criar o box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x="Metric", y="Score", data=df_melted, palette="Set2")

# Adicionar ttulo e ajustar rtulos
plt.title("Box Plot of Model Performance Metrics")
plt.xlabel("Metric")
plt.ylabel("Score")
plt
```
